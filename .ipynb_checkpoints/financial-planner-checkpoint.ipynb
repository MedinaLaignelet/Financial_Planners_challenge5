### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
from dotenv import load_dotenv
load_dotenv("Keys.env")
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
today = pd.Timestamp("2024-01-22", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to "1Day" for Alpaca API
timeframe = "1Day"

# Set start and end datetimes last weekday
start_date = pd.Timestamp("2024-01-19", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2024-01-19", tz="America/New_York").isoformat()


# Get current closing prices for SPY and AGG
ticker_data = api.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

ticker_data.head(20)



```

```python
# Separate ticker data
AGG = ticker_data[ticker_data['symbol'] == 'AGG'].drop('symbol', axis=1)
SPY = ticker_data[ticker_data["symbol"] == 'SPY'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
ticker_data = pd.concat([AGG,SPY], axis=1, keys=["AGG","SPY"])

# Preview DataFrame
ticker_data.head()
```

```python
# Pick AGG and SPY close prices
agg_close_price = float(ticker_data["AGG"]["close"])
spy_close_price = float(ticker_data["SPY"]["close"])
# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_spy_value = my_spy * spy_close_price
my_agg_value = my_agg * agg_close_price
total_stockbond_value = my_spy_value + my_agg_value
total_stockbond_value
# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
print(f"The total crypto value of both AGG and SPY is ${total_stockbond_value: 0.2f}")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of 3 years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2018-05-01', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-05-01', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# Set timeframe to "1Day"
timeframe = "1Day"

# Set the ticker information
tickers = ["AGG", "SPY"]

# HINT: Set "limit" to at least 10000 so all ticker rows are captured from get_bars()
df_histdata = api.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df

# Display sample data
df_histdata.head()

# Reorganize the DataFrame
# Separate ticker data
AGG = df_histdata[df_histdata['symbol'] == 'AGG'].drop('symbol', axis=1)
SPY = df_histdata[df_histdata["symbol"] == 'SPY'].drop('symbol', axis=1)

# Concatenate the ticker DataFrames
df_histdata = pd.concat([AGG,SPY], axis=1, keys=["AGG","SPY"])

# Display sample data
df_histdata.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
MC_dist = MCSimulation(
    portfolio_data = df_histdata,
    weights = [.40,.60],
    num_simulation = 1000,
    num_trading_days = 252 * 30
)
```

```python
# Printing the simulation input data
MC_dist.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_dist.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_dist.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_dist.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
even_tbl = MC_dist.summarize_cumulative_return()

# Print summary statistics
print(even_tbl)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(even_tbl[8]*20000,2)
ci_upper = round(even_tbl[9]*20000,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(even_tbl[8]*initial_investment,2)
ci_upper = round(even_tbl[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
